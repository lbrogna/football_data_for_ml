The prediction task I will tackle in this demonstration will be to forecast whether or not a given match will end in a draw.

I create a target boolean series that reflects this below.

This is just one of many potential questions that can be asked using this dataset, so users are encouraged to customize the target to their needs.

```python
target_draw = df['FTR'] == 'D'
target_draw
```

Here are two more ideas for targets. 

The first represents a 3-way categorical target for whether a match will end an a home win, an away win, or a draw. The second is a boolean target representing whether or not the total goals scored will exceed 2.5. 

I won't use these in this demo, but they're here to provide guidance in case you do want to use them or something similar.

```python
# 3-way target 
target_hda = df[[col for col in df.columns if 'FTR' in col]]
# over/under 2.5 goals target
sum_column = df['FTHG'] +  df['FTAG']
target_ou = sum_column > 2.5
```

I also need to prepare my dataset so that it makes predictions using only data that was collected before match start time. I'll get rid of the stats collected during the course of a match.

```python
during_match_cols = ['FTHG', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'HTR', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR']
df = df.drop(during_match_cols, axis=1)
```

I am also going to drop all the columns which contain the data for the closing odds. I don't want to use these because they are collected at odds close and betting stats, which is usually pretty close to the start of the match. With real data, it wouldn't be practical to wait until then to make a prediction.

```python
closing_odds_cols = ['B365CH' ,  'B365CD' ,  'B365CA' ,  'BWCH' ,  'BWCD' ,  'BWCA' ,  'IWCH' ,  'IWCD' ,  'IWCA' ,  'PSCH' ,  'PSCD' ,  'PSCA' ,  'WHCH' ,  'WHCD' ,  'WHCA' ,  'VCCH' ,  'VCCD' ,  'VCCA' ,  'MaxCH' ,  'MaxCD' ,  'MaxCA' ,  'AvgCH' ,  'AvgCD' ,  'AvgCA' ,  'B365C>2.5' ,  'B365C<2.5' ,  'PC>2.5' ,  'PC<2.5' ,  'MaxC>2.5' ,  'MaxC<2.5' ,  'AvgC>2.5' ,  'AvgC<2.5' ,  'AHCh' ,  'B365CAHH' ,  'B365CAHA' ,  'PCAHH' ,  'PCAHA' ,  'MaxCAHH' ,  'MaxCAHA' ,  'AvgCAHH' ,  'AvgCAHA']
df = df.drop(closing_odds_cols, axis=1)
```

There are some categorical data to deal with as well; teams, league, and season.

There are too many unique teams across all the leagues to warrant a one-hot encoding for the HomeTeam and AwayTeam features in my opinion. I'll just drop these.

I will one-hot encode the league and season however.

```python
df = df.drop(columns=['HomeTeam', 'AwayTeam'], axis=1)
df = pd.get_dummies(df, columns=['Div'])
df = pd.get_dummies(df, columns=['season'])
```

There is also a datetime column which I could extract days, months, and years from in order to one-hot encode them, but I think my league position columns should convey most of the information that those date columns could. For this demo I'll be dropping the Date column.

```python
df = df.drop(columns=['Date'], axis=1)
```

This is the point where one would reduce the feature count if they thought there were too many columns.

I've purposefully kept my dataset smaller than it could be, so I won't do this, but if during data processing the user included many historical columns, or looked back at a large number of historical matches, it is recommended that some sort of feature reduction method be implemented.

A few good methods of doing this would be: 
1. Aggregating the historical data by way of averaging historical statistics over a certain time span or spans.

2. Using PCA to reduce dimensionality on all columns or certain related columns (such as all match stats or all historical match stats).

3. Finding the importance of features to your target and removing those features with low importance. 

Methods 2 and 3 should be used with caution. 

PCA reduces dimensionality by representing linear variation between variables with less features. This means that along with the loss in information that you specify for your PCA (i.e. what percent of the variation will be lost when reducing the features), you will also lose any non-linear relationships between variables.

Similarly, methods of measuring importance are usually measurements of linear correlation between input features and the target. This means that non-linear relationships are not taken into account.

The data is almost ready to be split into sets for training, testing, and validation. Before that though, I'll make sure the dtypes are in order.

```python
# drop all rows containing non-numeric data
df = df[df.applymap(lambda x: isinstance(x, (int, float))).all(1)]

# drop same rows in targets
indices = target_hda.index.isin(df.index)

target_hda = target_hda[indices]
target_ou = target_ou[indices]

# cast types
df = df.astype(float)
```

It is also good practice to scale the data so that the values within every feature fall between the same numbers. This will speed up training of the model by eliminating arbitrary magnitudes which are less important than the actual variation within each feature.

```python
from sklearn.preprocessing import StandardScaler

# create a StandardScaler object
scaler = StandardScaler()

# fit the scaler to the data and transform the data
scaled_data = scaler.fit_transform(df)

# create a new dataframe from the scaled data
df = pd.DataFrame(scaled_data, columns=df.columns)
```

Now the data can be split into the necessary sets so that they can be used in a model.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, target_draw, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)

# print the shapes of the datasets to confirm they are split correctly
print("Input Train shape:", X_train.shape)
print("Input Validation shape:", X_val.shape)
print("Input Test shape:", X_test.shape)
```

Indices can be used to get the binary target's train, test, and validation sets.

```python
indices = target_draw.index.isin(y_train.index)
y_train = target_draw[indices]
indices = target_draw.index.isin(y_test.index)
y_test = target_draw[indices]
indices = target_draw.index.isin(y_val.index)
y_val = target_draw[indices]
print("Target train shape:", y_train.shape)
print("Target validation shape:", y_val.shape)
print("Target test shape:", y_test.shape)
```

Below I'm configuring a TPU. This is not strictly necessary, but it is reccomended if using Google Colab. Wrapping models in the TPU greatly speeds up processing times, so one should take advantage of this.

```python
import tensorflow as tf
import os

# Authenticate and configure the TPU
try:
  tpu = tf.distribute.cluster_resolver.TPUClusterResolver()
  print('Running on TPU ', tpu.cluster_spec().as_dict()['worker'])
except ValueError:
  raise BaseException('ERROR: Not connected to a TPU runtime')
  
tf.config.experimental_connect_to_cluster(tpu)
tf.tpu.experimental.initialize_tpu_system(tpu)
tpu_strategy = tf.distribute.experimental.TPUStrategy(tpu)
```

Here I define a simple NN to use with my binary target (draw/not draw). It includes dropout for the input layer. BN is also included after the first hidden layer.

```python
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

# Define the model architecture
def create_model(activation, dropout_rate, hidden_layer_size, num_hidden_layers, optimizer='adam'):
    model = keras.Sequential()
    model.add(Dropout(dropout_rate))
    for i in range(num_hidden_layers):
        if i == 0:
            model.add(Dense(hidden_layer_size, input_dim=X_train.shape[1], activation=activation))
            model.add(keras.layers.BatchNormalization())
        else:
            model.add(Dense(hidden_layer_size, activation=activation))
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

Next, a thorough hyperparameter optimization step is performed.

Warning: This will take a while to run depending on your number of features, even with the TPU. 

The user should customize all these parameters for their input and target, but especially 'hidden_layer_size' to better fit the number of features in your input.

```python
# Define the hyperparameter grid to search over
param_grid = {
    'activation': ['relu', 'LeakyReLU'],
    'dropout_rate': [0.5, 0.7, 0.9],
    'hidden_layer_size': [3, 5, 10],
    'num_hidden_layers' : [1, 2, 3]
}

# wrap the model creation in a tpu strategy so the code runs faster
with tpu_strategy.scope():
  model = KerasClassifier(build_fn=create_model, verbose=0)
  grid_search = GridSearchCV(model, param_grid, cv=5, verbose=3)

# Fit the GridSearchCV object to the training data
grid_search.fit(X_train, y_train, validation_data=(X_val, y_val))

# Print the best hyperparameters found by GridSearchCV
print('Best hyperparameters: ', grid_search.best_params_)
```

Below is a method that will dynamically adjust the learning rate as epochs increase. I will begin learning with a learning rate of .001, which is fairly standard for ADAM.

```python
import numpy as np
from keras.callbacks import LearningRateScheduler

def step_decay_schedule(initial_lr=1e-3, decay_factor, step_size):
    '''
    Wrapper function to create a LearningRateScheduler with step decay schedule.
    '''
    def schedule(epoch):
        return initial_lr * (decay_factor ** np.floor(epoch/step_size))
    
    return LearningRateScheduler(schedule)
```

Here I initialize the model within the TPU strategy with the optimized hyperparameters. I also initialize the learning rate schedule to begin at 0.001, multiplied by 0.75 after every epoch. This setup performed the best during my own testing, but will need to be changed when using different inputs and/or targets. It took relatively few epochs (only 10) for the model to reach a point where the validation loss stopped decreasing, but this will vary according to how the user has set up the rest of the experiment.

```python
with tpu_strategy.scope():
  model = create_model(activation='relu', dropout_rate=0.5, optimizer='adam', hidden_layer_size=1, num_hidden_layers=2)

lr_sched = step_decay_schedule(initial_lr=0.001, decay_factor=0.75, step_size=1)

history = model.fit(X_train, y_train, callbacks=[lr_sched], epochs=10, batch_size=32, validation_data=(X_val, y_val))
```

Now evaluating the trained model on the test set:

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test loss: {test_loss:.4f}')
print(f'Test accuracy: {test_accuracy:.4f}')
```

As something extra to test out the model with a real world application, I want to simulate bets on match draws informed by model outputs.

I'll use the test set for this as well, and create a new dataframe containing the values I'll need to complete this task.

First, I unscale the data to get the odds for a draw in every match of the test set.

```python
unscaled_data = scaler.inverse_transform(X_test)
test_df = pd.DataFrame(unscaled_data, columns=df.columns)
draw_odds = test_df['AvgD']
draw_odds
```

Next, I want to add the target (draw/ no draw) so I can evaluate each simulated bet as a win or a loss.

```python
data = {'odds': np.array(draw_odds),
        'outcome': np.array(y_test)}
bet_df = pd.DataFrame(data)
bet_df
```

The next step is to add the model outputs for every match in the test set. The model outputs represent model's expected probability that a draw will occur in that given match. Because I will want to calculate the expected value of each bet, I add both the model output and 1 - model output (p, and 1-p).

```python
bet_df['model_p'] = model.predict(X_test)
bet_df['model_p-1'] = 1 - bet_df['model_p']
bet_df
```

With these columns, we can add the model's expected value for each bet using a basic probability equation.

```python
bet_df['ev'] = (bet_df['model_p'] * bet_df['odds']) + (bet_df['model_p-1'] * -1)
bet_df
```

Finally, I'll create a function that can use these columns to simulate bets and winnings given a minimum expected wager value. A bet will only be placed if its expected value exceeds this number. For all qualifying wagers, a stake of $1 is placed. The function then prints the total money spent on bets, and the total winnings.

```python
def simulate_wagers(df, min_ev):
  qualifying_wagers = df[df['ev'] >= min_ev]
  qualifying_wagers['result'] = qualifying_wagers.apply(lambda row: -1 if not row['outcome'] else row['odds'], axis=1)
  total_spent = len(qualifying_wagers)
  total_winnings = round(qualifying_wagers['result'].sum(), 2)
  gain_or_loss = round((total_winnings - total_spent), 2)
  print('For minimum expected value = ' + str(min_ev) + ':\n')
  print('Total wagered: $' + str(total_spent))
  print('Total winnings: $' + str(total_winnings))
  print('Total gain/loss: $' + str(gain_or_loss) + '\n')
```

Now simulating wagers for several different minimum expected values:

```python
simulate_wagers(bet_df, min_ev=1)
simulate_wagers(bet_df, min_ev=1.1)
simulate_wagers(bet_df, min_ev=1.3)
simulate_wagers(bet_df, min_ev=1.5)
simulate_wagers(bet_df, min_ev=2)
simulate_wagers(bet_df, min_ev=3)
```
